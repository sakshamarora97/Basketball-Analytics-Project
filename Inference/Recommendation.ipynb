```python
league = 'NCAAW1+'
season = '22-23'
TEAM_ID = "618ade0180945020656b09d9"
LAST_X_GAMES = 10
NUM_RECOMMENDATIONS = 10
MIN_ATTEMPTS_VALIDITY = 3
```

```python
stars = pd.read_pickle(f'data/processed_data/{league}_{season}_stars.pickle')
```

```python
relevant_columns =["shot_id","team_id","player_id","game_id","date","Zone","Stat","Shot_type","ptr","epps","made"]
```

```python
stars = stars[relevant_columns]
```

```python
stars["apps"] = stars["ptr"]*stars["made"]
```

```python
stars
```

```python
def get_team_last_games(stars_df,team_id,last_x_games):
    master = stars_df.loc[stars_df.team_id==team_id].sort_values(by="date",ascending=False).reset_index(drop=True)
    all_games = master[["game_id","date"]].drop_duplicates().reset_index(drop=True)
    if all_games.shape[0]>=last_x_games:
        rel_games = all_games.iloc[0:last_x_games]["game_id"]
        master = master.loc[master.game_id.isin(rel_games)].reset_index(drop=True)
        print(f"Number of games selected {rel_games.shape[0]}")
        return master
    elif all_games.shape[0]<last_x_games:
        print(f"Number of games selected {all_games.shape[0]}")        
        return master
```

```python
relevant_stars = get_team_last_games(stars,ObjectId(TEAM_ID),LAST_X_GAMES)
```

```python
relevant_stars
```

```python
a = relevant_stars.groupby(["player_id","Zone","Shot_type","ptr"],as_index=False)[["epps","apps"]].mean().rename(columns={"epps":"mean_epps","apps":"mean_apps"})
b= relevant_stars.groupby(["player_id","Zone","Shot_type","ptr"],as_index=False)["shot_id"].count().rename(columns={"shot_id":"num_attempts"})
stats = b.merge(a,on=["player_id","Zone","Shot_type","ptr"],how="left")
stats
```

```python
#if close to 0, implies no insight
stats["pps_diff"] = stats["mean_epps"] - stats["mean_apps"]
#higher implies opportunity and lower apps due to randomness
#lower implies chance shots and higher apps due to luck
```

```python
c = stars.groupby(["Zone","Shot_type","ptr"],as_index=False)[["epps","apps"]].mean().rename(columns={"epps":"mean_epps","apps":"mean_apps"})
d= stars.groupby(["Zone","Shot_type","ptr"],as_index=False)["shot_id"].count().rename(columns={"shot_id":"overall_num_attempts"})
overall_stats = c.merge(d,on=["Zone","Shot_type","ptr"],how="left").rename(columns= {"mean_epps":"overall_mean_epps","mean_apps":"overall_mean_apps"})
overall_stats
```

```python
all_stats = stats.merge(overall_stats,on=["Zone","Shot_type","ptr"],how="left")
```

```python
all_stats
```

```python
take_more = all_stats.loc[all_stats.num_attempts>=MIN_ATTEMPTS_VALIDITY].sort_values(by="pps_diff",ascending=False).head(NUM_RECOMMENDATIONS)
take_less = all_stats.loc[all_stats.num_attempts>=MIN_ATTEMPTS_VALIDITY].sort_values(by="pps_diff",ascending=True).head(NUM_RECOMMENDATIONS)
```

```python
take_more
```

```python
take_less
```

```python
params = pd.DataFrame({"league":[league],"season":[season],"TEAM_ID":[TEAM_ID],"LAST_X_GAMES":[LAST_X_GAMES],"NUM_RECOMMENDATIONS":[NUM_RECOMMENDATIONS],"MIN_ATTEMPTS_VALIDITY":[MIN_ATTEMPTS_VALIDITY]})
```

```python
params
```

```python
writer =  pd.ExcelWriter(f'Inference/{league}_{TEAM_ID}_recommendations.xlsx',engine="xlsxwriter")
a= params
b= take_more.drop(columns=["overall_mean_epps",	'overall_mean_apps', 'overall_num_attempts'])
c = take_less.drop(columns=["overall_mean_epps",	'overall_mean_apps', 'overall_num_attempts'])
a.to_excel(writer, sheet_name='Parameters',startrow=1, header=False, index=False)
b.to_excel(writer, sheet_name='High Quality Low Conversion',startrow=1, header=False, index=False)
c.to_excel(writer, sheet_name='Low Quality High Conversion',startrow=1, header=False, index=False)
```

```python
dfs = [a,b,c]
```

```python
for i,sheet in enumerate(writer.sheets):
    worksheet = writer.sheets[sheet]

    # Get the dimensions of the dataframe.
    (max_row, max_col) = dfs[i].shape

    # Create a list of column headers, to use in add_table().
    column_settings = []
    for header in dfs[i].columns:
        column_settings.append({'header': header})

    # Add the table.
    worksheet.add_table(0, 0, max_row, max_col - 1, {'columns': column_settings})

    # Make the columns wider for clarity.
    worksheet.set_column(0, max_col - 1, 12)

    # Close the Pandas Excel writer and output the Excel file.
writer.close()
```
